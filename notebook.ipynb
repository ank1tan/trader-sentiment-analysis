# Trader Performance vs Market Sentiment Analysis

```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fg = pd.read_csv('fear_greed_index.csv')
tr = pd.read_csv('hyperliquid_data.csv')

print(fg.head())
print(tr.head())
```

## Data Cleaning

```

fg.columns = fg.columns.str.lower()
tr.columns = tr.columns.str.lower()

fg['date'] = pd.to_datetime(fg['date']).dt.date

# try common timestamp column names
time_col = [c for c in tr.columns if 'time' in c][0]
tr['time'] = pd.to_datetime(tr[time_col], errors='coerce')
tr['date'] = tr['time'].dt.date

df = tr.merge(fg[['date','classification']], on='date', how='left')

df['win'] = df['closedpnl'] > 0
```

## Analysis

```

print("Avg PnL by sentiment")
print(df.groupby('classification')['closedpnl'].mean())

print("Win rate by sentiment")
print(df.groupby('classification')['win'].mean())
```

## Charts

```

sns.barplot(data=df, x='classification', y='closedpnl')
plt.show()

sns.barplot(data=df, x='classification', y='win')
plt.show()
```
